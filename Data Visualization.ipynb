# RANDOM WALK
to generate data for a random walk and
then use matplotlib to create a visually appealing representation of the
generated data. A random walk is a path that has no clear direction but is
determined by a series of random decisions, each of which is left entirely to
chance. You might imagine a random walk as the path an ant would take if
it had lost its mind and took every step in a random direction.
Random walks have practical applications in nature, physics, biology,
chemistry, and economics. For example, a pollen grain floating on a drop
of water moves across the surface of the water because it is constantly being
pushed around by water molecules. Molecular motion in a water drop is
random, so the path a pollen grain traces out on the surface is a random
walk. The code were about to write models many real-world situations.

```python
from random import choice

class RandomWalk():
    """To generate random walks according to random decisions."""
    
    def __init__(self, number_of_points=5000):
        """To initialize the number of walks."""
        self.number_of_points = number_of_points
        
        # All walk starts at (0,0)
        self.x_values = [0]
        self.y_values = [0]
        
    def fill_walk(self):
        """Calculate all the points in walk."""
        
        # Keeping taking steps until the walk reaches the desired length.
        while len(self.x_values) < self.number_of_points:
            
            # Decide which direction to go and how far to go in that direction.
            x_direction = choice([1, -1])
            x_distance = choice([0, 1, 2, 3, 4])
            x_step = x_direction * x_distance
            
            y_direction = choice([1, -1])
            y_distance = choice([0, 1, 2, 3, 4])
            y_step = y_direction * y_distance
            
            # Reject moves that go nowhere.
            if x_step == 0 and y_step == 0:
                continue
                
            # Calculate the next x and y values.
            next_x = self.x_values[-1] + x_step
            next_y = self.y_values[-1] + y_step
            
            self.x_values.append(next_x)
            self.y_values.append(next_y)
            
rw = RandomWalk()
rw.fill_walk()
# print(rw.x_values)
# print(rw.y_values)
plt.scatter(rw.x_values, rw.y_values, s=15)
```

# Random Walk [Practise]


```python
from random import choice

class RandomWalk():
    """To model a class that will count walks or steps of any object that are completely random."""
    
    def __init__(self, total_number_of_walks=1000):
        """To initialize the total number of walks."""
        
        self.total_number_of_walks = total_number_of_walks
        
        # to store x points, whether object has moved to right or left.
        # current value will be 0 in the list.
        self.x_values = [0]
        
        # to store y points, whether object has moved to top or bottom.
        # current value will be 0 in the list.
        self.y_values = [0]
        
    def filling_walk(self):
        """Checking the direction and how much distance will that object covers for a particular point."""
        
        while len(self.x_values) < self.total_number_of_walks:
            """To check whether x_values are less than total number of walks to iterate till end."""
            
            x_direction = choice([1, -1]) # The choice will select a random number, 1 -> right and -1 -> left.
            x_distance = choice([0, 1, 2, 3, 4]) # Selection is random too for how much object moved in that direction.
            x_step = x_direction * x_distance
            
            y_direction = choice([1, -1]) # The choice will select a random number, 1 -> top and -1 > bottom.
            y_distance = choice([0, 1, 2, 3, 4]) # Selection is random too for how much object moved in that direction.
            y_step = y_direction * y_distance
            
            # if both x and y are 0, it means the object has not moved and it is useless to plot that point.
            if x_step == 0 and y_step == 0:
                continue
            
            # to get the next values we add the steps to last element stored in both x an y.
            next_x_value = self.x_values[-1] + x_step
            next_y_value = self.y_values[-1] + y_step
            
            self.x_values.append(next_x_value)
            self.y_values.append(next_y_value)

rw = RandomWalk()
rw.filling_walk()
plt.scatter(rw.x_values, rw.y_values, c='indigo', s=15)
plt.savefig('randomwalks', bbox_inches='tight')
```
